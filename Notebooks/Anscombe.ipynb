```python
## Variance
```

```python
df = anscombe.groupby('dataset')['x'].var()

x_axis = df.index
y_axis = df.values

fig = plt.figure( figsize=( 5,6 ) )


graph = sns.barplot( x=x_axis, y=y_axis , color='#f59e42' )
graph.set(ylim=(0,14))

plt.show()
```

```python
df = anscombe.groupby('dataset')['y'].var()

x_axis = df.index
y_axis = df.values

fig = plt.figure( figsize=( 5,6 ) )


graph = sns.barplot( x=x_axis, y=y_axis , color='#f59e42' )
graph.set(ylim=(0,10))

plt.show()
```

```python
## Correlations

correlations = dict()

for dataset in anscombe['dataset'].unique():
    corr = df1['x'].corr(df1['y'])
    correlations[dataset] = corr


graph = sns.scatterplot( x=list(correlations.keys()), y=list(correlations.values()), color='#f59e42' , s = 100)
graph.set(ylim=(-1,1))
plt.show()
```
