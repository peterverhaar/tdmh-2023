```python
import requests
from tdmh import *
from collections import Counter
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import nltk
from nltk import word_tokenize,pos_tag
```

```python
url = 'https://www.gutenberg.org/cache/epub/174/pg174.txt'
response = requests.get(url)
full_text = response.text
full_text = remove_pg_boilerplate(full_text).lower()
```

## Lemmatisation

```python
lemmatiser = WordNetLemmatizer()
pos = pos_tag(words)

lemmatised_words = []

for i,word in enumerate(words):
    word = word.lower()
    
    posTag = ptb_to_wordnet( pos[i][1] )
    
    if re.search( r'\w+' , posTag , re.IGNORECASE ):
        lemma = lemmatiser.lemmatize( words[i] , posTag )
        print( f'{word} => {lemma}' )
        lemmatised_words.append(lemma)
    else:
        print( f'{word} => {word}' )
        lemmatised_words.append(word)
```

```python
freq = Counter(lemmatised_words)

x_values = []
y_values = []
for word,count in freq.most_common(25):
    y_values.append(word)
    x_values.append(count)


fig = plt.figure( figsize=( 7 , 8 ) )

sns.barplot( x= x_values , y= y_values , color = '#9e291c'  )
plt.yticks(fontsize=20)

plt.xticks(rotation= 90)


plt.show()
```

## Linguistic constructions

```python


def make_pattern(pos_tags):
    pattern = ''
    tags = re.split( r'\s+' , pos_tags )
    for t in tags:
        pattern += r'\w+##' + t + ' '
    return pattern.strip()

def remove_tags(tagged_text):
    pattern = ''
    tokens = re.split( r'\s+' , tagged_text )
    for t in tokens:
        pattern += t[:t.index('#')] + ' '
    return pattern.strip()
```

```python



pattern = make_pattern('DT RB\w? JJ\w? NN\w?')



full_text = re.sub( r'\n' , ' ' , full_text)
sentences = sent_tokenize(full_text)

for sent in sentences:
    words = word_tokenize(sent)
    words = remove_punctuation(words)
    pos = pos_tag(words)
    
    tagged_sentence = ''

    for p in pos:
        tagged_sentence += p[0]+'##'+p[1] + ' '
        
    hits = re.findall( pattern , tagged_sentence)
    for h in hits:        
        print(f'{remove_tags(h)}')
        
```
